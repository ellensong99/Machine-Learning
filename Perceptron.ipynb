### Perceptron

```python
import pandas as pd
import io
# read data from local
dfX = pd.read_csv(io.BytesIO(uploaded['spambase_X.csv']), header=None)
dfy = pd.read_csv(io.BytesIO(uploaded['spambase_y.csv']), header=None)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def perceptron(X, y, w, b, max_pass):
  mistake = [0 for i in range(0, max_pass)]
  for t in range(0, max_pass):
    for i in range(0, X.shape[1]):
      xi = np.array(X[i])
      if y[i] * (np.dot(xi, np.array(w)) + b) <= 0:
        w = w + y[i] * X[i]
        b = b + y[i]
        mistake[t] = mistake[t] + 1
  return w, b, mistake

y = dfy[0].values.tolist()
max_pass = 500
b = 0
d = dfX.shape[0]
w = [0 for i in range(0, d)]
w, b, mistake = perceptron(dfX, y, w, b, max_pass)
```

```python
from matplotlib.pyplot import figure
figure(figsize=(10, 8), dpi=80)
plt.plot(mistake)
plt.title("Number of Mistakes w.r.t Number of Passes")
plt.ylabel("Number of Mistakes")
plt.xlabel("Number of Passes")
```
